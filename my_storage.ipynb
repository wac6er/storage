```python
import yfinance as yf

def get_stock_percent_change(ticker_symbol):
    stock = yf.Ticker(ticker_symbol)
    history = stock.history(period="15d")
    prev_close = history.iloc[0]['Close']
    last_close = history.iloc[1]['Close']
    percent_change = ((last_close - prev_close) / prev_close) * 100

    return percent_change, history

ticker = "EXR"
extra_change,history = get_stock_percent_change(ticker)

ticker = "PSA"
public_change, history1= get_stock_percent_change(ticker)
history1
```

```python
import pandas as pd
import re

def extract_size(size_str):
    dimensions = re.findall(r'(\d+\'?)', size_str)
    if len(dimensions) == 2:
        width, length = dimensions
        return float(width.replace("'", "")), float(length.replace("'", ""))
    else:
        return None, None

def extra_space(lifestorage_df_1,progress_extraspace_df_1,lifestorage_df_2,progress_extraspace_df_2):

  lifestorage_df_1['unique_id'] = lifestorage_df_1['Store_ID'].astype(str) + '_' + lifestorage_df_1['Unit_ID'].astype(str)
  lifestorage_df_2['unique_id'] = lifestorage_df_2['Store_ID'].astype(str) + '_' + lifestorage_df_2['Unit_ID'].astype(str)

  progress_extraspace_df_1['unique_id'] = progress_extraspace_df_1['Store_ID'].astype(str) + '_' + progress_extraspace_df_1['Unit_ID'].astype(str)
  progress_extraspace_df_2['unique_id'] = progress_extraspace_df_2['Store_ID'].astype(str) + '_' + progress_extraspace_df_2['Unit_ID'].astype(str)

  lifestorage_df_1 = lifestorage_df_1[lifestorage_df_1['unique_id'].isin(lifestorage_df_2['unique_id'])]
  lifestorage_df_2 = lifestorage_df_2[lifestorage_df_2['unique_id'].isin(lifestorage_df_1['unique_id'])]

  progress_extraspace_df_1 = progress_extraspace_df_1[progress_extraspace_df_1['unique_id'].isin(progress_extraspace_df_2['unique_id'])]
  progress_extraspace_df_2 = progress_extraspace_df_2[progress_extraspace_df_2['unique_id'].isin(progress_extraspace_df_1['unique_id'])]

  lifestorage_df_1.drop('unique_id', axis=1, inplace=True)
  lifestorage_df_2.drop('unique_id', axis=1, inplace=True)
  print(len(lifestorage_df_1))
  print(len(lifestorage_df_2))

  progress_extraspace_df_1.drop('unique_id', axis=1, inplace=True)
  print(len(progress_extraspace_df_1))
  progress_extraspace_df_2.drop('unique_id', axis=1, inplace=True)
  print(len(progress_extraspace_df_2))

  results = []

  for df in [lifestorage_df_1, progress_extraspace_df_1, lifestorage_df_2, progress_extraspace_df_2]:
      df = df.drop_duplicates(subset='Unit_ID', keep='first')
      df['Units_Rented'] = df['Total'] - df['Available']
      df[['Width', 'Length']] = df['Size '].apply(extract_size).apply(pd.Series)
      df['Square_Footage'] = df['Width'] * df['Length']
      quarterly_revenue_current = 3 * (df['Units_Rented'] * df['Current_Price']).sum()
      quarterly_revenue_initial = 3 * (df['Units_Rented'] * df['Initial_Price']).sum()
      total_available_sqft = (df['Square_Footage'] * df['Available']).sum()
      total_rented_sqft = (df['Square_Footage'] * df['Units_Rented']).sum()
      total_sqft = total_available_sqft + total_rented_sqft
      percentage_available = (total_available_sqft / total_sqft) * 100
      percentage_rented = (total_rented_sqft / total_sqft) * 100
      results.append([quarterly_revenue_current, quarterly_revenue_initial, percentage_available, percentage_rented])

  total_sqft_lifestorage_1 = (results[0][2] / 100) * (results[0][2] + results[0][3])
  total_sqft_extraspace_1 = (results[1][2] / 100) * (results[1][2] + results[1][3])
  total_sqft_lifestorage_2 = (results[2][2] / 100) * (results[2][2] + results[2][3])
  total_sqft_extraspace_2 = (results[3][2] / 100) * (results[3][2] + results[3][3])

  overall_total_sqft_1 = total_sqft_lifestorage_1 + total_sqft_extraspace_1
  overall_total_sqft_2 = total_sqft_lifestorage_2 + total_sqft_extraspace_2

  percentage_change_lifestorage = ((total_sqft_lifestorage_2 - total_sqft_lifestorage_1) / total_sqft_lifestorage_1) * 100
  percentage_change_extraspace = ((total_sqft_extraspace_2 - total_sqft_extraspace_1) / total_sqft_extraspace_1) * 100

  weighted_percentage_change = ((total_sqft_lifestorage_1 / overall_total_sqft_1) * percentage_change_lifestorage) + ((total_sqft_extraspace_1 / overall_total_sqft_1) * percentage_change_extraspace)

  changes = {
      'Metric': ['Quarterly Revenue (Current Price)', 'Quarterly Revenue (Initial Price)', 'Percentage Available', 'Percentage Rented'],
      'Snapshot 1 (LifeStorage)': results[0],
      'Snapshot 1 (ExtraSpace)': results[1],
      'Snapshot 2 (LifeStorage)': results[2],
      'Snapshot 2 (ExtraSpace)': results[3],
      'Nominal Change (LifeStorage)': [results[2][i] - results[0][i] for i in range(4)],
      'Percent Change (LifeStorage)': [((results[2][i] - results[0][i]) / results[0][i]) * 100 for i in range(4)],
      'Nominal Change (ExtraSpace)': [results[3][i] - results[1][i] for i in range(4)],
      'Percent Change (ExtraSpace)': [((results[3][i] - results[1][i]) / results[1][i]) * 100 for i in range(4)]
  }

  results_df = pd.DataFrame(changes)
  results_df=results_df[results_df['Metric']!='Quarterly Revenue (Initial Price)']

  revenue_difference_current_lifestorage = (results[2][0] - results[0][0])/3
  revenue_difference_initial_lifestorage = (results[2][1] - results[0][1])/3
  sqft_difference_lifestorage = total_sqft_lifestorage_2 - total_sqft_lifestorage_1

  revenue_difference_current_extraspace = (results[3][0] - results[1][0])/3
  revenue_difference_initial_extraspace = (results[3][1] - results[1][1])/3
  sqft_difference_extraspace = total_sqft_extraspace_2 - total_sqft_extraspace_1

  # Construct the corrected results dataframe
  corrected_results_data = {
      'Storage Provider': ['LifeStorage', 'ExtraSpace'],
      'Revenue Difference (Current Price)': [revenue_difference_current_lifestorage, revenue_difference_current_extraspace],
      'Revenue Difference (Initial Price)': [revenue_difference_initial_lifestorage, revenue_difference_initial_extraspace],
      'Square Feet Difference': [sqft_difference_lifestorage, sqft_difference_extraspace]
  }

  corrected_results_df_new = pd.DataFrame(corrected_results_data)

  percent_revenue_change = sum([results_df['Nominal Change (LifeStorage)'][0],results_df['Nominal Change (ExtraSpace)'][0]])/sum([results_df['Snapshot 1 (LifeStorage)'][0],results_df['Snapshot 1 (ExtraSpace)'][0]])


  return corrected_results_df_new, str(percent_revenue_change)[:6]+'%', str(weighted_percentage_change)[:6]+'%',

#lifestorage_df_1 = pd.read_csv('/content/lifestorage.csv')
#progress_extraspace_df_1 = pd.read_csv('/content/Extra_Space1.csv')
#lifestorage_df_2 = pd.read_csv('/content/life_storage_2.csv')
#progress_extraspace_df_2 = pd.read_csv('/content/Extra_Space_2.csv')
#lifestorage_df_3 = pd.read_csv('/content/life_storage_3.csv')
#progress_extraspace_df_3 = pd.read_csv('/content/Extra_Space_3.csv')
#lifestorage_df_4 = pd.read_csv('/content/life_storage_4.csv')
#progress_extraspace_df_4 = pd.read_csv('/content/Extra_Space_4.csv')
#lifestorage_df_5 = pd.read_csv('/content/life_storage_4.csv')
#progress_extraspace_df_5 = pd.read_csv('/content/Extra_Space_5.csv')
lifestorage_df_6 = pd.read_csv('/content/life_storage_16.csv')
progress_extraspace_df_6 = pd.read_csv('/content/Extra_Space_16.csv')
lifestorage_df_7 = pd.read_csv('/content/life_storage_17.csv')
progress_extraspace_df_7 = pd.read_csv('/content/Extra_Space_17.csv')


#corrected_results_df_new, rev_change, sq_change = extra_space(lifestorage_df_1,progress_extraspace_df_1,lifestorage_df_2,progress_extraspace_df_2)
#corrected_results_df_new_1, rev_change1, sq_change1 = extra_space(lifestorage_df_2,progress_extraspace_df_2,lifestorage_df_3,progress_extraspace_df_3)
#corrected_results_df_new_2, rev_change1, sq_change1 = extra_space(lifestorage_df_3,progress_extraspace_df_3,lifestorage_df_4,progress_extraspace_df_4)
#corrected_results_df_new_3, rev_change1, sq_change1 = extra_space(lifestorage_df_4,progress_extraspace_df_4,lifestorage_df_5,progress_extraspace_df_5)
corrected_results_df_new_3, rev_change1, sq_change1 = extra_space(lifestorage_df_6,progress_extraspace_df_6,lifestorage_df_7,progress_extraspace_df_7)
```

```python
import pandas as pd
import re

def public_storage(df1,df2):
  df1_cleaned = df1.drop_duplicates(subset=['Unit_ID', 'Store_ID'])
  df2_cleaned = df2.drop_duplicates(subset=['Unit_ID', 'Store_ID'])

  def extract_square_feet(size_string):
      dimensions = re.findall(r"(\d+(\.\d+)?)", size_string)
      if len(dimensions) == 2:
          width, length = float(dimensions[0][0]), float(dimensions[1][0])
          return width * length
      else:
          return None

  df1_cleaned['Square_Feet'] = df1_cleaned['Unit_Size'].apply(extract_square_feet)
  df2_cleaned['Square_Feet'] = df2_cleaned['Unit_Size'].apply(extract_square_feet)

  merged_df = pd.merge(df1_cleaned, df2_cleaned, on=['Unit_ID', 'Store_ID'], how='inner', suffixes=('_df1', '_df2'))

  merged_df['Count_Difference_Corrected'] = merged_df['Count_df1'] - merged_df['Count_df2']
  merged_df['Revenue_Difference_Current_Price_Corrected'] = merged_df['Count_Difference_Corrected'] * merged_df['Current_Price_df1']
  merged_df['Revenue_Difference_Initial_Price_Corrected'] = merged_df['Count_Difference_Corrected'] * merged_df['Initial_Price_df1']
  merged_df['Square_Feet_Difference_Corrected'] = merged_df['Count_Difference_Corrected'] *-1* merged_df['Square_Feet_df1']

  corrected_results = {
      'Total Revenue Difference (Current Price)': merged_df['Revenue_Difference_Current_Price_Corrected'].sum(),
      'Total Revenue Difference (Initial Price)': merged_df['Revenue_Difference_Initial_Price_Corrected'].sum(),
      'Total Square Feet Difference': merged_df['Square_Feet_Difference_Corrected'].sum()
  }

  increase_count = (merged_df['Count_Difference_Corrected'] < 0).sum()
  decrease_count = (merged_df['Count_Difference_Corrected'] > 0).sum()

  corrected_results_df = pd.DataFrame([corrected_results])
  return corrected_results_df

#df1 = pd.read_csv('/content/Public_Storage_1.csv')
#df2 = pd.read_csv('/content/Public_Storage_2.csv')
#df3 = pd.read_csv('/content/Public_Storage_3.csv')
#df4 = pd.read_csv('/content/Public_Storage_4.csv')
#df5 = pd.read_csv('/content/Public_Storage_5.csv')
df6 = pd.read_csv('/content/Public_Storage_16.csv')
df7 = pd.read_csv('/content/Public_Storage_17.csv')

#corrected_results_df = public_storage(df1,df2)
#corrected_results_df_1 = public_storage(df2,df3)
#corrected_results_df_2 = public_storage(df3,df4)
#corrected_results_df_3 = public_storage(df4,df5)
corrected_results_df_3 = public_storage(df6,df7)
```
