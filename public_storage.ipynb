```python
driver = webdriver.Chrome(options=chrome_options)

from uszipcode import SearchEngine

# Initialize the search engine
search = SearchEngine()  # set `simple_zipcode=False` for richer data, but it's slower

# Retrieve all ZIP codes
zipcodes = search.by_population(lower=1, sort_by='population', ascending=False, returns=1000)

# Extract the top 1000 ZIP codes by population
top_1000_zip = [zipcode.zipcode for zipcode in zipcodes]
df = []
locations = top_1000_zip
i = 0
for location in locations:
  try:
    i = i+1
    print(i)
 # try:
    url = 'https://www.publicstorage.com/self-storage-search?location='+location

    driver.get(url)
   # time.sleep(3)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    soup = driver.page_source
    soup = str(BeautifulSoup(soup, "lxml"))
    spl = soup.split('data-gtmdata')

    for ii in range(len(spl)-1):
        unit_id = spl[ii].split('data-unit-id="')[-1].split('"')[0]
        store_id = spl[ii+1].split('storeId&quot;:&quot;')[1].split('&')[0]
        size = spl[ii+1].split('name&quot;:&quot;')[1].split('&')[0]
        initial_price = spl[ii+1].split('listprice&quot;:')[1].split(',')[0]
        current_price = spl[ii+1].split('saleprice&quot;:')[1].split(',')[0]
        min_price = spl[ii+1].split('minPrice&quot;:')[1].split(',')[0]
        count = spl[ii+1].split('count&quot;:')[1].split(',')[0]
        features = spl[ii+1].split('filterClasses&quot;:&quot;')[1].split('&')[0]

        sp = soup.split('"postalCode":"')
        df_loc = []
        for t in range(len(sp)-1):
          postal_code = sp[t+1].split('"')[0]
          store_code = sp[t].split('"storeID":"')[-1].split('"')[0]
          df_loc.append({
              "Postal_Code":postal_code,
              "Store_Code":store_code
          })
        df_loc = pd.DataFrame(df_loc)
        for aa in range(len(df_loc)):
          if store_id == df_loc['Store_Code'][aa]:
            address = df_loc['Postal_Code'][aa]
            break
          else:
            address = 'NaN'
        df.append({
            'Unit_ID':unit_id,
            'Store_ID':store_id,
            'Unit_Size':size,
            'Current_Price':current_price,
            'Min_Price':min_price,
            'Initial_Price':initial_price,
            'Count':count,
            'Features':features,
            'Location':location,
            'Postal_Code':address
            })
    pd.DataFrame(df).drop_duplicates().to_csv('progress.csv')
    print(len(pd.DataFrame(df).drop_duplicates(subset = ['Store_ID','Unit_ID'])))

  except:
    print('ERROR')
    time.sleep(5)
# except:
 #   print('ERROR')
  #  driver.quit()
   # time.sleep(10)
    #driver = webdriver.Chrome(options=chrome_options)
```
